```python
# Define the weights for each performance division (5 sets of weights)
weights_sets = [
    [0, 5, 10, 15, 20, 25],  # %0-5_Points
    [0, 5, 10, 15, 20, 25],  # %6-10_Points
    [0, 5, 10, 15, 20, 25],  # %11-15_Points
    [0, 5, 10, 15, 20, 25],  # %16-20_Points
    [0, 5, 10, 15, 20, 25]   # %21-25_Points
]

# Calculate the weighted score for each row in the DataFrame
performance_columns = [
    '%0-5_Points', '%6-10_Points', '%11-15_Points', '%16-20_Points', '%21-25_Points',
    '%0-5_Points.1', '%6-10_Points.1', '%11-15_Points.1', '%16-20_Points.1', '%21-25_Points.1',
    '%0-5_Points.2', '%6-10_Points.2', '%11-15_Points.2', '%16-20_Points.2',
    '%0-5_Points.3', '%6-10_Points.3', '%11-15_Points.3', '%16-20_Points.3',
    '%0-5_Points.4', '%6-10_Points.4', '%11-15_Points.4', '%16-20_Points.4'
]

# Create a new DataFrame to store the scores
df_scores = df_all[['District_Name', 'SCHOOL', 'Gender']].copy()

# Calculate the weighted score for each row and add to the new DataFrame
for i, division in enumerate(performance_columns):
    weight_points = weights_sets[i % 5]  # Get the appropriate set of weights for each performance division
    df_scores[division] = df_all[division].mul(weight_points[i % 6] / 100)
    
# Sum all the performance divisions to get the weighted score
df_scores['Weighted_Score'] = df_scores[performance_columns].sum(axis=1).round(2)

# Print the DataFrame with the added score column
df_scores
```

```python
# Define the number of top performing schools you want to retrieve overall
top_n = 20

# Sort the DataFrame by 'Weighted_Score' in descending order
df_scores_sorted = df_scores.sort_values(by='Weighted_Score', ascending=False)

# Select the top 20 rows which represent the top performing schools overall
top_performing_schools_overall = df_scores_sorted.head(top_n)

# Reset the index to reindex the resulting DataFrame
top_performing_schools_overall.reset_index(drop=True, inplace=True)

# Print the DataFrame containing the top performing schools overall
print(top_performing_schools_overall)
```

```python
#Presenting the top performing schools using a map
import pandas as pd
import geocoder
import folium

# Function to get latitude and longitude for a given address using Geocoder
def get_coordinates(address):
    location = geocoder.osm(address)
    return location.latlng

# Create a new DataFrame to store the top performing schools with latitude and longitude
top_performing_schools_df = pd.DataFrame(columns=['SCHOOL', 'District_Name', 'Latitude', 'Longitude', 'Weighted_Score'])

# Loop through each district and retrieve the top performing schools
for district in top_performing_schools_overall['District_Name'].unique():
    # Get the top performing schools for the current district
    top_schools_district = top_performing_schools_overall[top_performing_schools_overall['District_Name'] == district].nlargest(10, 'Weighted_Score')
    
    # Add latitude and longitude information for each school
    top_schools_district['Coordinates'] = top_schools_district['SCHOOL'].apply(get_coordinates)
    
    # Filter out rows with None values in the Coordinates column
    top_schools_district = top_schools_district.dropna(subset=['Coordinates'])
    
    # Create separate lists for Latitude and Longitude
    latitude_list = []
    longitude_list = []
    for coordinates in top_schools_district['Coordinates']:
        latitude_list.append(coordinates[0])
        longitude_list.append(coordinates[1])
    
    # Assign the lists to the new DataFrame
    top_schools_district['Latitude'] = latitude_list
    top_schools_district['Longitude'] = longitude_list
    
    # Append the top performing schools for the current district to the new DataFrame
    top_performing_schools_df = pd.concat([top_performing_schools_df, top_schools_district[['SCHOOL', 'District_Name', 'Latitude', 'Longitude', 'Weighted_Score']]])

# Create a map centered on the mean latitude and longitude of all schools
map_top_performing_schools = folium.Map(location=[top_performing_schools_df['Latitude'].mean(), top_performing_schools_df['Longitude'].mean()], zoom_start=10)

# Loop through the top performing schools DataFrame and add markers to the map
for index, row in top_performing_schools_df.iterrows():
    school_name = row['SCHOOL']
    latitude = row['Latitude']
    longitude = row['Longitude']
    weighted_score = row['Weighted_Score']
    
    # Create a tooltip for the marker with school information
    tooltip = f"{school_name}<br>Weighted Score: {weighted_score}"
    
    # Add a marker for each school to the map
    folium.Marker(location=[latitude, longitude], popup=school_name, tooltip=tooltip).add_to(map_top_performing_schools)

# Display the map
map_top_performing_schools
```
